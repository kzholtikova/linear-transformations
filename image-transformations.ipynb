```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('src/penguin.jpg')
height, width, ch = img.shape
plt.imshow(cv.bitwise_not(img))
```

```python
rotated_img =  cv.warpAffine(img, cv.getRotationMatrix2D(((width-1)/2.0,(height-1)/2.0),-45,1),(width, height))
plt.imshow(cv.bitwise_not(rotated_img))
```

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
# image as if viewed from a different angle or position
perspective_img = cv.warpPerspective(rotated_img, cv.getPerspectiveTransform(pts1,pts2), (300,300))
plt.imshow(cv.bitwise_not(perspective_img))
```

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
# preserves lines and parallelism in the image
affine_img = cv.warpAffine(img, cv.getAffineTransform(pts1,pts2), (width, height))
plt.imshow(cv.bitwise_not(affine_img))
```
