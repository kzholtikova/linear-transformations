```python
import numpy as np
import object 
```

```python
pyramid = np.array([
    [0, 0, 0],  # Base point 1
    [1, 0, 0],  # Base point 2
    [0.5, 1, 0],  # Base point 3 (skewed)
    [0, 0, 1],  # Top point
    [0, 0, 0],  # Back to base point 1
    [1, 0, 0],  # Base point 2
    [0, 0, 1],  # Top point
    [1, 0, 0],  # Base point 2
    [0.5, 1, 0],  # Base point 3
    [0, 0, 1],  # Top point
    [0.5, 1, 0],  # Base point 3
    [0, 0, 0]  # Base point 1
])
pyramid = object.Object(pyramid, "Pyramid")
```

```python
pyramid.plot('red')
```
